```python
import pandas as pd

allc=pd.read_csv('clustered_df_all.csv')
econ=pd.read_csv('clustered_df_economy.csv')
covid=pd.read_csv('clustered_df_covid.csv')
```

```python
allc.head(10)
```

```python
econ=econ[['country_code','cluster']]
covid=covid[['country_code','cluster']]
```

```python
allc=allc.rename(columns={'cluster':'101'})
econ=econ.rename(columns={'cluster':'102'})
covid=covid.rename(columns={'cluster':'103'})
allc=allc.rename(columns={'gdp':'1','tourism':'2','unemployment':'3','gdp_growth':'4','debt_rate':'5','vaccination':'6', 'death_rate':'7'})
```

```python
data=pd.merge(allc, econ, on=['country_code'], how='outer')
data=pd.merge(data, covid, on=['country_code'], how='outer')
```

```python
data.to_csv('clusters.csv', index=False)
```
